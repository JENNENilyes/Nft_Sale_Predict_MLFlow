### Functions for Models Metrics

```python
def evaluate(estimator, X_tr, y_tr):
    '''
    This is to evalute the best models and outputs are the metrics and matrix
    estimator: model,
    X_tr: X_train,
    y_tr: y_training set
    '''
#    predicts the models
    tr_preds = estimator.predict(X_tr)

    
    print("Training Scores:")
    print(f"Train Accuracy: {accuracy_score(y_tr, tr_preds)}")
    print(f"Train precision: {precision_score(y_tr, tr_preds)}")
    print(f"Train recall: {recall_score(y_tr, tr_preds)}")
    print(f"Train f1 score: {f1_score(y_tr, tr_preds)}")

    return(accuracy_score(y_tr, tr_preds),precision_score(y_tr, tr_preds),recall_score(y_tr, tr_preds),f1_score(y_tr, tr_preds))
```

#### K  Nearest Neighbor Classifier

```python
# Fonction qui permet d'entraner le modle Elasticnet
import warnings

def train_knn(train_conv_nft_df, y_train,test_nft_df,y_test,n_neighbors=1, data_path=data_path):
    """
    Entranement du modle Elastic net en changeant les paramtres alpha et l1_ratio
    """
    warnings.filterwarnings("ignore")
    
    
    with mlflow.start_run():
      
        knn = KNeighborsClassifier(n_neighbors = n_neighbors )
        knn.fit(train_conv_nft_df, y_train)
        knn_predict = knn.predict(test_nft_df)
        Accuracy,precision,recall,f1_score=evaluate(knn, test_nft_df, y_test)

        mlflow.log_param(key="n_neighbors", value= n_neighbors)
        #mlflow.log_params( {"alpha": alpha, "l1_ratio": l1_ratio} )
        mlflow.log_metric(key="Accuracy", value= Accuracy)
        mlflow.log_metric(key="precision", value= precision)
        mlflow.log_metric(key="recall", value= recall)
        mlflow.log_metric(key="f1_score", value= f1_score)

        #mlflow.log_metrics({"rmse" : rmse, "mae", mae, "r2": r2})
        mlflow.log_artifact(data_path)
        print(" Save to: {}".format(mlflow.get_artifact_uri() ))
        mlflow.sklearn.log_model(knn,"knn_model")


    return Accuracy, precision, recall,f1_score
```

#### Support Vector Classifier

```python
from sklearn.svm import  SVC
def train_SVC(train_conv_nft_df, y_train,test_nft_df,y_test, data_path=data_path):
    """
    Entranement du modle Elastic net en changeant les paramtres alpha et l1_ratio
    """
    warnings.filterwarnings("ignore")
    
    
    with mlflow.start_run():
        
        
        svc_model = SVC()
        svc_model.fit(train_conv_nft_df, y_train)

        y_predict = svc_model.predict(test_nft_df)
        Accuracy,precision,recall,f1_score=evaluate(svc_model, test_nft_df, y_test)



        #mlflow.log_param(key="n_neighbors", value= n_neighbors)
        #mlflow.log_params( {"alpha": alpha, "l1_ratio": l1_ratio} )
        mlflow.log_metric(key="Accuracy", value= Accuracy)
        mlflow.log_metric(key="precision", value= precision)
        mlflow.log_metric(key="recall", value= recall)
        mlflow.log_metric(key="f1_score", value= f1_score)

        #mlflow.log_metrics({"rmse" : rmse, "mae", mae, "r2": r2})
        mlflow.log_artifact(data_path)
        print(" Save to: {}".format(mlflow.get_artifact_uri() ))
        mlflow.sklearn.log_model(svc_model,"SVC_model")


    return Accuracy, precision, recall,f1_score
```

#### MLP Classifier

```python
from sklearn.svm import  SVC
def train_MLP(train_conv_nft_df, y_train,test_nft_df,y_test, data_path=data_path,C=1 , gamma=0.1):
    """
    Entranement du modle Elastic net en changeant les paramtres alpha et l1_ratio
    """
    warnings.filterwarnings("ignore")
    
    
    with mlflow.start_run():
        
        
        #best model
        best_svc = SVC(C=C , gamma=gamma)
        best_svc.fit(train_conv_nft_df, y_train)
        y_pred = best_svc.predict(test_nft_df)

        Accuracy,precision,recall,f1_score=evaluate(best_svc, test_nft_df, y_test)



        mlflow.log_param(key="C", value= C)
        mlflow.log_param(key="gamma", value= gamma)

        #mlflow.log_params( {"alpha": alpha, "l1_ratio": l1_ratio} )
        mlflow.log_metric(key="Accuracy", value= Accuracy)
        mlflow.log_metric(key="precision", value= precision)
        mlflow.log_metric(key="recall", value= recall)
        mlflow.log_metric(key="f1_score", value= f1_score)

        #mlflow.log_metrics({"rmse" : rmse, "mae", mae, "r2": r2})
        mlflow.log_artifact(data_path)
        print(" Save to: {}".format(mlflow.get_artifact_uri() ))
        mlflow.sklearn.log_model(best_svc,"SVC_model")


    return Accuracy, precision, recall,f1_score
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
def train_tree(train_conv_nft_df, y_train,test_nft_df,y_test, data_path=data_path,random_state=42):
    """
    Entranement du modle Elastic net en changeant les paramtres alpha et l1_ratio
    """
    warnings.filterwarnings("ignore")
    
    
    with mlflow.start_run():
        
        
        nft_tree = DecisionTreeClassifier(random_state=42)
        nft_tree.fit(train_conv_nft_df, y_train)
        y_pred = nft_tree.predict(test_nft_df)

        Accuracy,precision,recall,f1_score=evaluate(nft_tree, test_nft_df, y_test)



        mlflow.log_param(key="random_state", value= random_state)
        #mlflow.log_params( {"alpha": alpha, "l1_ratio": l1_ratio} )
        mlflow.log_metric(key="Accuracy", value= Accuracy)
        mlflow.log_metric(key="precision", value= precision)
        mlflow.log_metric(key="recall", value= recall)
        mlflow.log_metric(key="f1_score", value= f1_score)

        #mlflow.log_metrics({"rmse" : rmse, "mae", mae, "r2": r2})
        mlflow.log_artifact(data_path)
        print(" Save to: {}".format(mlflow.get_artifact_uri() ))
        mlflow.sklearn.log_model(nft_tree,"tree_model")


    return Accuracy, precision, recall,f1_score
```
